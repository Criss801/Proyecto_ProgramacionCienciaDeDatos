```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar el archivo
df = pd.read_csv('california_housing.csv')

# 3. Medidas estadísticas con NumPy
ingresos = df['MedInc'].values
media_ingreso = np.mean(ingresos)
mediana_ingreso = np.median(ingresos)
desviacion_ingreso = np.std(ingresos)

print(f"Media del ingreso: {media_ingreso:.2f}")
print(f"Mediana del ingreso: {mediana_ingreso:.2f}")
print(f"Desviación Estándar: {desviacion_ingreso:.2f}")
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['HouseAge'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución de la Edad de las Casas en California')
plt.xlabel('Años')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Seleccionamos las variables
x = df['MedInc'].values
y = df['AveRooms'].values # Relación entre ingresos y tamaño de la casa

# 6. Implementar regresión lineal con numpy.polyfit
# Grado 1 significa línea recta: y = mx + b
m, b = np.polyfit(x, y, 1)

# 7. Graficar
plt.figure(figsize=(10,6))
plt.scatter(x, y, alpha=0.5, label='Datos reales', s=10)
plt.plot(x, m*x + b, color='red', linewidth=3, label='Línea de Regresión')
plt.title('Relación entre Ingreso Medio y Promedio de Habitaciones')
plt.xlabel('Ingreso Medio (Escala 10k USD)')
plt.ylabel('Promedio de Habitaciones')
plt.legend()
plt.show()
```
